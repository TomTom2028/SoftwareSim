```python
import numpy as np
import pandas as pd

#df1 = pd.read_excel("./data/anon_WMSPick20250101_03.xlsx") # Slechte data: timestamps kloppen niet
df2 = pd.read_excel("./data/anon_WMSPick20250104_06.xlsx")
df3 = pd.read_excel("./data/anon_WMSPick20250107_09.xlsx")
#df4 = pd.read_excel("./data/anon_WMSPick20250110_12.xlsx") # idem
df5 = pd.read_excel("./data/anon_WMSPick20250113_15.xlsx")
df6 = pd.read_excel("./data/anon_WMSPick20250116_19.xlsx")

dfs=[df2,df3,df5,df6]
df=pd.concat(dfs)
df = df[df['Out in'] != 'Outgoing']
df
```

```python
df =df[df['Location code'] == "OSR"]
```

```python
df['Requester user code'].value_counts().sort_index()
```

```python
four_user_code = df[df['Requester user code'] == 72]
four_user_code
```

```python
four_user_code.columns.tolist()
```

```python
import numpy as np

delta_last_change_dts = []
prev_time = None
for row in four_user_code.itertuples():
    if prev_time is None:
        prev_time = row[5]
    else:
        delta = (row[5] - prev_time).total_seconds()
        prev_time = row[5]
        if delta > 100:
            continue
        if delta < 0.5:
            continue
        delta_last_change_dts.append(delta)


delta_last_change_dts = np.array(delta_last_change_dts)
delta_last_change_dts

```

```python
# create a histogram of the delta_last_change_dts
import matplotlib.pyplot as plt
plt.hist(delta_last_change_dts, bins=50)
plt.xlabel('Delta last change (seconds)')
plt.ylabel('Frequency')
plt.title('Histogram of Delta Last Change Times')
plt.show()
```

```python
# print average
print("Average delta last change time:", np.mean(delta_last_change_dts))
```

```python
average_time_for_package = np.mean(delta_last_change_dts) # in seconds
packages_per_second = 1 / average_time_for_package
packages_per_hour = packages_per_second * 3600
print(f"Average packages per hour: {packages_per_hour:.2f}")
```
