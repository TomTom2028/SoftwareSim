```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation
mu, sigma = 1.39, 0.07

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute the PDF
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, label=f"Avg: {(mu *3.6):.2f} km/h")
plt.title('Walking speed of picker (m/s)')
plt.xlabel('Speed (m/s)')
plt.ylabel('Chance')
plt.grid(True)
plt.legend()
plt.show()
```
