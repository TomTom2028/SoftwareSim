```python
import json
def load_normal_json(filepath:str):
    with open(filepath, 'r') as reader:
        json_obj = json.load(reader)
    return json_obj
```

```python
import numpy as np
bins = np.linspace(0, 80, 60)
bin_width = bins[1] - bins[0]
bin_centers = bins[:-1] + bin_width / 2
```

```python
currentJson = load_normal_json('./output_tests/two_vlms_one_and_a_half_lift.json')
histedDeltas, _ = np.histogram(currentJson["delta_times"], bins=bins)
avgDelta = sum(currentJson["delta_times"]) / len(currentJson["delta_times"])
```

```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
plt.bar(bin_centers, histedDeltas / len(currentJson["delta_times"]), width=bin_width, label=f"avg: {avgDelta:.2f}")
plt.legend()
plt.title(f"Histogram of Delta Times for {currentJson['name']}")
plt.xlabel("Delta Time (seconds)")
plt.ylabel("Frequency")
plt.show()
```
